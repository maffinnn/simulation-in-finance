```python
def retrieve_stock_history(ticker_id: str, auto_adjust: bool, start_date: str, end_date: str) -> pd.DataFrame:
    ticker = yf.Ticker(ticker_id)
    ticker.options
    # opt = ticker.option_chain('2023-11-06')
    # print(opt)
    return ticker.history(auto_adjust = auto_adjust, start = start_date, end = end_date)
```

```python
START_DATE = '2023-08-09'
END_DATE = '2023-11-06'
lonza = retrieve_stock_history(ticker_id = "LONN.SW", auto_adjust = False, start_date = START_DATE, end_date = END_DATE)
sika = retrieve_stock_history(ticker_id = "SIKA.SW", auto_adjust = False, start_date = START_DATE, end_date = END_DATE)

display(lonza)
display(sika)
```

```python
lonza_array = np.array(lonza_close['Adj Close'])
sika_array = np.array(sika_close['Adj Close'])

cov_matrix = np.cov(lonza_array, sika_array)
print(f"Covariance matrix is:\n {cov_matrix}\n")
print(f"The shape is {np.shape(cov_matrix)}\n")

print(f"Correlation between the two var is {cov_matrix[0][1] / (cov_matrix[0][0] * cov_matrix[1][1]) ** 0.5}") # Correct

L = np.linalg.cholesky(cov_matrix)
print(f"The matrix after Cholesky decomposition is:\n {L}\n")

print(f"The multiplication of L and L transpose is:\n {np.dot(L, L.T)}\n") 
# Cannot use *
# Retrieve the original covariance matrix
```

## Covariance matrix

```python
arr_1 = np.array([0, 1, 3, 5])
arr_2 = np.array([1, 2, 8, 5])

cov_matrix = np.cov(arr_1, arr_2)
print(np.shape(cov_matrix))
print(cov_matrix)
```

```python
x = np.array([[0, 2], [5, 1], [2, 0]])
# 3 rows, 2 columns
print(x)
print(np.cov(x))
print(np.var(x[0]))
print(np.var(x[0], ddof = 1)) 
# Delta degrees of freedom = 1 for sample variance
# Now the variance matches the diagonal elements in the covariance
```

```python
x = [1.23, 2.12, 3.34, 4.5]  
y = [2.56, 2.89, 3.76, 3.95]  
    
# find out covariance with respect  
# rows  
cov_mat = np.stack((x, y), axis = 1)   

print(cov_mat)

print("shape of matrix x and y:",  
      np.shape(cov_mat))  
  
print("shape of covariance matrix:", 
      np.shape(np.cov(cov_mat)))  
  
print(np.cov(cov_mat))
```
